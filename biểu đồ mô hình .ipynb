```python
import matplotlib.pyplot as plt
import numpy as np

# Tên các mô hình
models = ['Random Forest', 'XGBoost', 'LightGBM']

# Các chỉ số đánh giá theo thứ tự: R², RMSE, MAE, Accuracy
r2_scores = [0.9981, 0.8730, 0.9790]
rmse_scores = [0.0430, 0.00904, 0.1429]
mae_scores = [0.0048, 0.0330334, 0.0075]
accuracy_scores = [0.7716, 0.6612, 0.8401]

x = np.arange(len(models))  # vị trí x cho các mô hình
width = 0.2  # độ rộng mỗi cột

# Vẽ biểu đồ
fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x - 1.5*width, r2_scores, width, label='R²', color='skyblue')
rects2 = ax.bar(x - 0.5*width, rmse_scores, width, label='RMSE', color='orange')
rects3 = ax.bar(x + 0.5*width, mae_scores, width, label='MAE', color='green')
rects4 = ax.bar(x + 1.5*width, accuracy_scores, width, label='Accuracy', color='purple')

# Thêm tiêu đề và nhãn
ax.set_ylabel('Giá trị')
ax.set_title('So sánh hiệu suất các mô hình')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Hiển thị số trên đầu cột
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}' if height < 1 else f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 6),
                    textcoords="offset points",
                    ha='center', va='bottom')

for bars in [rects1, rects2, rects3, rects4]:
    autolabel(bars)

plt.tight_layout()
plt.show()
```
